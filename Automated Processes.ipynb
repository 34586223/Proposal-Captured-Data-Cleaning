##### Importing Libraries 

```python
import pandas as pd

import numpy as np 

import re
```

#### Importing the Dataset

```python
df= pd.read_excel('Captured Cases Report V2_Captured Cases (37).xlsx')

df.head()
```

##### Removing Unwanted Columns

```python
df.columns
```

```python
df = df.drop(['POLICY CODE', 'Unnamed: 3', 'Unnamed: 7', 'CONVERTED BY', 'AUTHORISED BY', 'CAPTURED BY'], axis= 1)

df.head()
```

#### Rearranging Columns to Reporting Order

```python
df.columns
```

```python
df = df[['CLIENT', 'POLICY NO', 'PROPOSAL NO', 'STATUS', 'CAPTURE DATE', 'EFFECTIVE DATE', 'MODAL PREMIUM', 'ANNUALIZED PREMIUM',
          'PAY METHOD', 'PAYMENT FREQUENCY', 'CONVERSION DATE', 'AGENT', 'AGENT CODE', 'AGENT', 'AGENT JOIN DATE', 'AGENCY', 'UNIT']]

df.head()
```

```python
df['CAPTURE DATE'] = pd.to_datetime(df['CAPTURE DATE'], format='%d/%m/%Y')

df.head()
```

#### Changing Date Format to Short Date

```python
from datetime import datetime

df['CAPTURE DATE'] = pd.to_datetime(df['CAPTURE DATE'], format='%d/%m/%Y')

df['EFFECTIVE DATE'] = pd.to_datetime(df['EFFECTIVE DATE'], format='%d/%m/%Y')

df['CONVERSION DATE'] = pd.to_datetime(df['CONVERSION DATE'], format='%d/%m/%Y')

df['AGENT JOIN DATE'] = pd.to_datetime(df['AGENT JOIN DATE'], format='%d/%m/%Y')

df.head()
```

#### Converting Premium Values and Agent Codes to Numbers

```python
df['MODAL PREMIUM'] = pd.to_numeric(df['MODAL PREMIUM'])

df['ANNUALIZED PREMIUM'] = pd.to_numeric(df['ANNUALIZED PREMIUM'])

df[['AGENT CODE']] = df[['AGENT CODE']].apply(pd.to_numeric, errors='coerce')

df.head()
```

```python
# Leaving Null Values Empty

df = df.fillna('')

df.head()
```

#### Ordering Columns Using Capture Date

```python
df = df.sort_values(by='CAPTURE DATE', ascending=True)

df.head()
```

#### Removing Duplicate Value

```python
df.duplicated('PROPOSAL NO')
```

```python
df = df.drop_duplicates(subset= ['PROPOSAL NO'], keep= "last")

df.head()
```
